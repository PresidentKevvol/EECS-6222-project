```python
err_amount_array = np.load('sim_result_b.npy')
```

```python
# min and max channel error
min_p = 0.00
max_p = 0.15

# number of intervals
num_interval = 60

p_values = np.linspace(min_p, max_p, num_interval+1)
```

```python
m = 8
k = 223
err_amount_total = np.sum(err_amount_array, axis=1) / (m*k*100)
```

```python
plt.plot(p_values, err_amount_total)

plt.xlabel("Error rate of BSC p")
plt.ylabel("Portion of bits in decoded message\nthat differs from original")

plt.title("Error portion vs. Error rate\nin Channel Using Voyager Encoding Scheme")
```
